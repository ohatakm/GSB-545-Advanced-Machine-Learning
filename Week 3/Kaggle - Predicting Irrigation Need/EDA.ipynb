```python
import pandas as pd
import sweetviz as sv
from plotnine import *

# load data
train = pd.read_csv("train.csv")

train.head()
```

```python
# Sweetviz report
report = sv.analyze(train)
report.show_html("irrigation_eda_report.html")
```

```python
(
    ggplot(train, aes(x="Soil_Moisture", fill="Irrigation_Need"))
    + geom_histogram(bins=30, alpha=0.6, position="identity")
    + labs(
        title="Soil Moisture Distribution by Irrigation Need",
        x="Soil Moisture",
        y="Count"
    )
)
```

I used Sweetviz to get a quick overview of the dataset and understand how the features relate to irrigation need . The dataset has a mix of numerical and categorical variables and didn’t really have major issues like missing values.
From the report, soil moisture, rainfall, and temperature looked like the most important features since their distributions varied more across the irrigation levels. Soil moisture especially stood out because lower values seemed more tied to higher irrigation need.
One thing that stood out is that a lot of the features still have overlapping distributions between classes, so it doesn’t look like a super clean separation problem and might be harder to classify.
Overall, the patterns made it seem like models that can handle non-linear relationships would probably work better.
