#### 3.
I mainly tried two models: a Random Forest as my baseline and XGBoost with some tuning. I used the same cleaned and encoded dataset for both so I could compare them pretty directly.
The Random Forest was solid to start and gave decent results without much effort, but it didn’t really improve much even when I tried tweaking it. XGBoost ended up working better overall, especially after adjusting things like learning rate and depth.
I also tried using Optuna for tuning, but it was taking over an hour to run on the full dataset, so I ended up using a sample instead. It was still helpful for seeing how automated tuning works, but I didn’t fully use those results for my final model.
The improvement from Random Forest to XGBoost wasn’t huge, but it was noticeable. It felt more like small gains rather than a big jump, but still enough to make XGBoost the better option.

#### 4.
Overall, boosting worked better than bagging. Random Forest was easier and more stable, but XGBoost picked up patterns a bit better and gave slightly stronger results, so I would go with that.

For phase 2, I’d focus more on feature engineering and tuning rather than just trying different models. A lot of the features have overlap, so improving the inputs will probably matter more than switching models again.
I also noticed from other Kaggle notebooks that a lot of people were doing feature engineering, so that’s something I’d try. I’d create new features based on interactions, especially between things like soil moisture, rainfall, and temperature since those seemed most important. I’d also try more tuning with XGBoost (or using Optuna more efficiently) to see if I can get small improvements. Another thing I’d look into is class imbalance or separation, since the overlap might be hurting performance.
